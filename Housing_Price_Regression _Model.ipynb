```python
def detect_outliers(data):
    outliers = []
    data = sorted(data)
    q1 = np.percentile(data, 25)
    q2 = np.percentile(data, 50)
    q3 = np.percentile(data, 75)
    print(f"q1:{q1}, q2:{q2}, q3:{q3}")

    IQR = q3-q1
    lwr_bound = q1-(1.5*IQR)
    upr_bound = q3+(1.5*IQR)
    print(f"Lower bound: {lwr_bound}, Upper bound: {upr_bound}, IQR: {IQR}")

    for i in data:
        if (i<lwr_bound or i>upr_bound):
            outliers.append(i)
    len_outliers= len(outliers)
    print(f"Total number of outliers are: {len_outliers}")

    print(f"Total percentage of outlier is: {round(len_outliers*100/len(data),2)} %")
```

```python
# Determining IQR, Lower and Upper bound and number out outliers present in each of the continous numerical feature
for feature in continous_value_columns:
  print(feature,":")
  detect_outliers(raw_housing_data[feature])
  print("\n")
```

```python
from scipy.stats.mstats import winsorize

def treat_outliers_iqr(data):
    # Calculate the first and third quartiles
    q1, q3 = np.percentile(data, [25, 75])

    # Calculate the interquartile range (IQR)
    iqr = q3 - q1

    # Identify the outliers
    lower_bound = q1 - (1.5 * iqr)
    upper_bound = q3 + (1.5 * iqr)
    outliers = [x for x in data if x < lower_bound or x > upper_bound]

     # Treat the outliers (e.g., replace with the nearest quartile value)
    treated_data = [q1 if x < lower_bound else q3 if x > upper_bound else x for x in data]
    treated_data_int = [int(absolute) for absolute in treated_data]

    return treated_data_int
```

```python
#Passing all the feature one by one from the list of continous_value_feature in our above defined function for outlier treatment
for feature in continous_value_columns:
  raw_housing_data[feature]= treat_outliers_iqr(raw_housing_data[feature])
```

```python
# Rechecking the total number of outliers and its percentage present in our dataset.
for feature in continous_value_columns:
  print(feature,":")
  detect_outliers(raw_housing_data[feature])
  print("\n")
```

Null hypothesis: Means are same at different levels

Alternate Hypothesis: Means are different. Categorical variable does have impact on target variable.

Anova analysis.

F-statistic: A higher F-statistic indicates greater variance between group means relative to within-group variance.

p-value:

If p-value â‰¤ 0.05: There is strong evidence to reject the null hypothesis, indicating that at least one group mean is different.

If p-value > 0.05: There is insufficient evidence to reject the null hypothesis, indicating that the group means are not significantly different.

```python
#ANOVA is done to check if mean of target variable is same or different across different levels of an independent categorical variable

from statsmodels.formula.api import ols
import statsmodels.api as sm
```

```python
def vif_calc(df):
  vif= pd.Series([variance_inflation_factor(df.values,i) for i in range(df.shape[1])], index= df.columns)

  if vif.max()>5:
    print(vif[vif==vif.max()].index[0],'has been removed')

    df= df.drop(columns= [vif[vif==vif.max()].index[0]])

    return df
  else:
    print("No more multicollinearity present.")
    return df
```

# **Model-2 using Forward Selection**

Selection Criterion- Adjusted R2

```python
x=raw_housing_data.drop(columns=['sale_price'])
y=raw_housing_data['sale_price']
```

```python
def adjusted_r2_score(y_true, y_pred, n, k):
    """ Calculate the adjusted R-squared. """
    r2 = r2_score(y_true, y_pred)
    return 1 - (1 - r2) * (n - 1) / (n - k - 1)

def forward_selection(X_train, y_train):
    """ Perform forward selection to find the best model with the highest adjusted R-squared. """
    initial_features = X_train.columns.tolist()
    best_features = []
    best_adj_r2 = -np.inf
    current_adj_r2 = -np.inf

    n = len(y_train)

    while initial_features:
        adj_r2_with_candidate = []

        for feature in initial_features:
            features_to_try = best_features + [feature]
            model = LinearRegression().fit(X_train[features_to_try], y_train)
            y_pred = model.predict(X_train[features_to_try])
            adj_r2 = adjusted_r2_score(y_train, y_pred, n, len(features_to_try))
            adj_r2_with_candidate.append((adj_r2, feature))

        adj_r2_with_candidate.sort(reverse=True)
        best_new_adj_r2, best_new_feature = adj_r2_with_candidate[0]

        if best_new_adj_r2 > current_adj_r2:
            best_features.append(best_new_feature)
            current_adj_r2 = best_new_adj_r2
            initial_features.remove(best_new_feature)
        else:
            break

    model = LinearRegression().fit(X_train[best_features], y_train)
    return model, best_features

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=100)

# Perform forward selection on the training data
best_model, best_features = forward_selection(X_train, y_train)

# Evaluate the model on the test data
y_pred_train = best_model.predict(X_train[best_features])
y_pred_test = best_model.predict(X_test[best_features])

print("Best features:", best_features)
print("Training R-squared:", best_model.score(X_train[best_features], y_train))
print("Training Adjusted R-squared:", adjusted_r2_score(y_train, y_pred_train, len(y_train), len(best_features)))
print("Test R-squared:", best_model.score(X_test[best_features], y_test))
print("Test Adjusted R-squared:", adjusted_r2_score(y_test, y_pred_test, len(y_test), len(best_features)))
```
